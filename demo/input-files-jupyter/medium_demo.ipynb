```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

rows = []
for index in range(12):
    rows.append({
        "month": index + 1,
        "revenue": 100 + index * 12,
        "cost": 55 + index * 7,
    })

df = pd.DataFrame(rows)
df["profit"] = df["revenue"] - df["cost"]
df.to_csv("medium_sales.csv", index=False)
```

```python
%%time
loaded = pd.read_csv("medium_sales.csv")
summary = loaded[["revenue", "cost", "profit"]].sum()
print(summary)
```

```python
ax = loaded.plot(x="month", y=["revenue", "cost", "profit"], marker="o", title="Monthly sales")
ax.set_ylabel("amount")
plt.show()
```
