# Small demo notebook

Small example with Markdown, DataFrame display, HTML display, and a plot magic.

```python
%matplotlib inline
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML

values = pd.DataFrame({
    "day": ["Mon", "Tue", "Wed", "Thu"],
    "orders": [12, 18, 15, 22],
})

display(values)
display(HTML("<p><b>Total orders:</b> {}</p>".format(values["orders"].sum())))
```

```python
ax = values.plot(x="day", y="orders", kind="bar", legend=False, title="Orders by day")
ax.set_ylabel("orders")
plt.show()
```
